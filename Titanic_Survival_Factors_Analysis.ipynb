5. Необхідно створити єдиний стовпчик, що вказує кількість родичів для кожного пасажира на борту, замість:
- кількість братів/сестер або чоловіків/дружин на борту;
- кількість батьків або дітей на борту;

Булева ознака: стовпчик 'alone'=True, якщо пасажир подорожував один (без родичів на борту).        
Перевір, щоб для одиноких пасажирів новий стовпчик мав значення 0.      
Після створення нового стовпчика, дропни "sibsp", "parch", "alone".     
Результат представ у вигляді підрахунку кількості пропущених значень (NaN) у стовпці relatives.

```python
df = sns.load_dataset('titanic')

df['relatives'] = df['sibsp'] + df['parch']
df.loc[df['alone'] == True, 'relatives'] = 0

print(df.loc[df['alone'] == True, ['relatives']].head())

df = df.drop(columns=['sibsp', 'parch', 'alone'])

print("Кількість NaN у relatives:", df['relatives'].isna().sum())

```

6. Перевір, як часто зустрічається певна кількість родичів в новому стовпці.    
Результат представ у вигляді таблиці, побудованої з використанням групування та агрегації:

```python
relatives_stat = (
    df.groupby('relatives')
      .agg(passengers_count=('relatives', 'count'))
      .reset_index()
      .sort_values('relatives')
)
print(relatives_stat)
```

7. Використовуючи цикл заміни кількість родичів, що перевищує число 5(п'ять) на значення "above 5"(понад п'ять).       
Запиши значення в новий стовпчик ʼrelatives_categoryʼ.     
Результат представ у вигляді таблиці, побудованої з використанням групування та агрегації:

```python
df['relatives_category'] = df['relatives'].astype(object)
for i in df.index:
    if df.loc[i, 'relatives'] > 5:
        df.loc[i, 'relatives_category'] = 'above 5'
    else:
        df.loc[i, 'relatives_category'] = df.loc[i, 'relatives']
relatives_category_stat = (
    df.groupby('relatives_category')
      .agg(passengers_count=('relatives_category', 'count'))
      .reset_index()
)
print(relatives_category_stat)
```

8. Необхідно вивести на екран статистику по відсотковому показнику пасажирів з кількістю родичів більше 5 відносно загального числа пасажирів для кожного з міст посадки. Для цього:
- порахуй загальну кількість пасажирів в кожному з міст посадки,
- знайди число пасажирів з кількістю родичів більше 5 в кожному з міст посадки,
- поділи ці два стовчики між собою, перетворивши результат у відсотки (ціле число).        
Відобрази статистику порахованих показників, згруповану по містах посадки.

```python
total_by_town = (
    df.groupby('embark_town')
      .agg(total_passengers=('embark_town', 'count'))
)
above_5_by_town = (
    df[df['relatives'] > 5]
    .groupby('embark_town')
    .agg(passengers_above_5=('embark_town', 'count'))
)
town_stat = total_by_town.join(above_5_by_town, how='left')
town_stat['passengers_above_5'] = town_stat['passengers_above_5'].fillna(0)
town_stat['percent_above_5'] = (
    (town_stat['passengers_above_5'] / town_stat['total_passengers']) * 100
).astype(int)
town_stat = town_stat.reset_index()
print(town_stat)
```

9. Заповни відсутні значення віку медіаною.

```python
import pandas as pd
import seaborn as sns

# Ensure df is defined with the base dataset to allow filling age.
# This assumes that if df is not defined, we should start with a fresh dataset
# for this specific operation, as transformations for 'relatives' etc.
# do not impact the 'age' column's median or missing values directly for this step.
df = sns.load_dataset('titanic')

df['age'] = pd.to_numeric(df['age'], errors='coerce')
age_median = df['age'].median()
print("Медіана age:", age_median)
df['age'] = df['age'].fillna(age_median)
print("Кількість пропусків після заповнення:", df['age'].isna().sum())
print("Мінімальний вік:", df['age'].min())
print("Перші значення age:")
print(df['age'].head())
```

10. Створи новий стовпець, де вік представлено категорією, замість числа (наприклад: до 14 років, 14-34 роки, 35-59 років, 60 і більше років). Виконай задачу з використанням користувацької функції. Осіб з невідомим віком познач відповідно.

```python
def age_category(age):
    if pd.isna(age):
        return 'невідомо'
    elif age < 14:
        return 'до 14 років'
    elif 14 <= age <= 34:
        return '14-34 роки'
    elif 35 <= age <= 59:
        return '35-59 років'
    else:
        return '60 і більше років'
df['age_category'] = df['age'].apply(age_category)
print(df[['age', 'age_category']].head(10))
print(df.groupby('age_category').agg(passengers_count=('age_category', 'count')).reset_index())
```

11. Перевір, в якій віковій категорії була найвища смертність.     
Для цього рекомендується перетворити стовпець 'alive' в булевий тип.    
 Потім підрахувати загальну кількість пасажирів та кількість тих, хто не вижив.      
 Потім обчисли відносний показниках для кожної категорії.

```python
df['alive_bool'] = df['alive'].map({'yes': True, 'no': False})
mortality_by_age = (
    df.groupby('age_category')
      .agg(
          total_passengers=('alive_bool', 'count'),
          dead_passengers=('alive_bool', lambda x: (~x).sum())
      )
      .reset_index()
)
mortality_by_age['mortality_percent'] = (
    mortality_by_age['dead_passengers'] / mortality_by_age['total_passengers'] * 100
).round(2)
print(mortality_by_age)
max_row = mortality_by_age.loc[mortality_by_age['mortality_percent'].idxmax()]
print(
    f"Найвища смертність була у віковій категорії: {max_row['age_category']} "
    f"— {max_row['mortality_percent']}%"
)
```

*Бонусне завдання*         :          
Підготуй розгорнуту статистику смертності по категорії віку, класу квитка, рівню каюти та кількості родичів.      
Які фактори, на твою думку, найбільше пов'язані з рівнем смерності? (наприклад: найбільша смертність у відсотковому значенні спостерігається серед вікової групи ... класу квитка.... при наявності ... родичів та для рівня каюти.... Фактор ... має найвищий вплив на смертність)

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset('titanic').copy()
# relatives
df['relatives'] = df['sibsp'] + df['parch']
df.loc[df['alone'] == True, 'relatives'] = 0
# age -> fillna median
df['age'] = df['age'].fillna(df['age'].median())
# age_category
def age_category(age):
    if pd.isna(age):
        return 'невідомо'
    elif age < 14:
        return 'до 14 років'
    elif 14 <= age <= 34:
        return '14-34 роки'
    elif 35 <= age <= 59:
        return '35-59 років'
    else:
        return '60 і більше років'

df['age_category'] = df['age'].apply(age_category)
# alive -> bool
df['alive_bool'] = df['alive'].map({'yes': True, 'no': False})
# death flag
df['dead'] = ~df['alive_bool']
# deck with missing as separate category
df['deck_stat'] = df['deck'].astype(object)
df['deck_stat'] = df['deck_stat'].fillna('Unknown')
mortality_age = (
    df.groupby('age_category')
      .agg(
          total_passengers=('dead', 'count'),
          dead_passengers=('dead', 'sum')
      )
      .reset_index()
)

mortality_age['mortality_percent'] = (
    mortality_age['dead_passengers'] / mortality_age['total_passengers'] * 100
).round(2)

mortality_age = mortality_age.sort_values('mortality_percent', ascending=False)
mortality_age
mortality_class = (
    df.groupby('class')
      .agg(
          total_passengers=('dead', 'count'),
          dead_passengers=('dead', 'sum')
      )
      .reset_index()
)

mortality_class['mortality_percent'] = (
    mortality_class['dead_passengers'] / mortality_class['total_passengers'] * 100
).round(2)

mortality_class = mortality_class.sort_values('mortality_percent', ascending=False)
mortality_class
mortality_deck = (
    df.groupby('deck_stat')
      .agg(
          total_passengers=('dead', 'count'),
          dead_passengers=('dead', 'sum')
      )
      .reset_index()
)

mortality_deck['mortality_percent'] = (
    mortality_deck['dead_passengers'] / mortality_deck['total_passengers'] * 100
).round(2)

mortality_deck = mortality_deck.sort_values('mortality_percent', ascending=False)
mortality_deck
mortality_relatives = (
    df.groupby('relatives')
      .agg(
          total_passengers=('dead', 'count'),
          dead_passengers=('dead', 'sum')
      )
      .reset_index()
)

mortality_relatives['mortality_percent'] = (
    mortality_relatives['dead_passengers'] / mortality_relatives['total_passengers'] * 100
).round(2)

mortality_relatives = mortality_relatives.sort_values('mortality_percent', ascending=False)
mortality_relatives
max_age = mortality_age.loc[mortality_age['mortality_percent'].idxmax()]
max_class = mortality_class.loc[mortality_class['mortality_percent'].idxmax()]
max_deck = mortality_deck.loc[mortality_deck['mortality_percent'].idxmax()]
max_relatives = mortality_relatives.loc[mortality_relatives['mortality_percent'].idxmax()]
print("Найвища смертність за віком:")
print(max_age)
print()

print("Найвища смертність за класом квитка:")
print(max_class)
print()

print("Найвища смертність за рівнем каюти:")
print(max_deck)
print()

print("Найвища смертність за кількістю родичів:")
print(max_relatives)
summary = pd.DataFrame({
    'factor': ['age_category', 'class', 'deck', 'relatives'],
    'worst_group': [
        max_age['age_category'],
        max_class['class'],
        max_deck['deck_stat'],
        max_relatives['relatives']
    ],
    'mortality_percent': [
        max_age['mortality_percent'],
        max_class['mortality_percent'],
        max_deck['mortality_percent'],
        max_relatives['mortality_percent']
    ]
})

summary = summary.sort_values('mortality_percent', ascending=False)
summary
print(
    f"Найбільша смертність у відсотковому значенні за віком спостерігається "
    f"у групі '{max_age['age_category']}' і становить {max_age['mortality_percent']}%."
)

print(
    f"За класом квитка найбільша смертність спостерігається у класі "
    f"'{max_class['class']}' і становить {max_class['mortality_percent']}%."
)

print(
    f"За рівнем каюти найбільша смертність спостерігається для рівня "
    f"'{max_deck['deck_stat']}' і становить {max_deck['mortality_percent']}%."
)

print(
    f"За кількістю родичів найбільша смертність спостерігається у пасажирів "
    f"з кількістю родичів '{max_relatives['relatives']}' і становить {max_relatives['mortality_percent']}%."
)

```
